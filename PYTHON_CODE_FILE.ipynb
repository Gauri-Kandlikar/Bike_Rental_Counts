```python
#Check for outliers of continuous variables  using boxplot
sns.boxplot(data=df1[['temp','windspeed','hum']])
fig=plt.gcf()
fig.set_size_inches(8,8)

##from the graph we can say that outliers are present in windspeed and humidity.
```

```python
#creating datafram
wh=pd.DataFrame(df1[['windspeed','hum']])
#columns for outliers
cnames=['windspeed','hum']
for i in cnames:
    q75,q25=np.percentile(wh.loc[:,i],[75,25]) # Divide data into 75 percentile and 25 percentile.
    iqr=q75-q25 #Inter quantile range
    min=q25-(iqr*1.5) #inner fence
    max=q75+(iqr*1.5) #outer fence
    wh.loc[wh.loc[:,i]<min,:i]=np.nan  #Replace with NA
    wh.loc[wh.loc[:,i]>max,:i]=np.nan  #Replace with NA







```

```python
pd.isnull(wh).sum()
```

```python
#Imputating the outliers by mean 
wh['windspeed']=wh['windspeed'].fillna(wh['windspeed'].mean())
wh['hum']=wh['hum'].fillna(wh['hum'].mean())
```

```python
###to verify
pd.isnull(wh).sum()
```

```python
#Replacing the imputated windspeed
df1['windspeed']=df1['windspeed'].replace(wh['windspeed'])

#Replacing the imputated humidity
df1['hum']=df1['hum'].replace(wh['hum'])
df1.head(5)
```

```python
##load the required libraries:-
from sklearn import preprocessing,metrics,linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
x=df1[['season','yr','mnth','holiday','weekday','workingday','weathersit','temp','hum','windspeed']]
y=df1['cnt']




x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)

#Reset train index values:-
x_train.reset_index(inplace=True)
y_train=y_train.reset_index()

# Reset train index values:-
x_test.reset_index(inplace=True)
y_test=y_test.reset_index()

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
print(y_train.head())
print(y_test.head())



```

```python
#categorical attributes
cat_attributes=['season','holiday','workingday','weathersit','yr']
#numerical attributes
num_attributes=['temp','windspeed','hum','mnth','weekday']
```

```python
#Decoding the training attributes:-
train_vars=x_train[['season','holiday','workingday','weathersit','yr','temp','windspeed','hum','mnth','weekday']]
test_vars=x_test[['season','holiday','workingday','weathersit','yr','temp','windspeed','hum','mnth','weekday']]
#To get dummy variables to encode the categorical features to numeric
train_encoded_attributes=pd.get_dummies(train_vars,columns=cat_attributes)
print('Shape of transfomed dataframe::',train_encoded_attributes.shape)
train_encoded_attributes.head(5)
```

```python
y_train
```

```python
##replacing dummmy encoded vars with the train dataset
x_train=train_encoded_attributes
y_train=y_train

```

# Building a Linear Regression Model

```python
clf=LinearRegression()
fitted=clf.fit(x_train,y_train)
fitted
```

```python
lr=clf.score(x_train,y_train)
print('Accuracy of the model :',lr)
print('Model coefficients :',clf.coef_)
print('Model intercept value :',clf.intercept_)
```

```python
#Decoding th test attributes:-
#To get dummy variables to encode the categorical features to numeric:-
test_encoded_attributes=pd.get_dummies(test_vars,columns=cat_attributes)
print('Shape of transformed dataframe :',test_encoded_attributes.shape)
test_encoded_attributes.head(5)
```

# Model Evaluation

```python
from sklearn import metrics
#Root mean square error and mean absolute error scores:-
import math
#Root mean square error 
rmse=math.sqrt(metrics.mean_squared_error(y_test,y_pred))
#Mean absolute error
mae=metrics.mean_absolute_error(y_test,y_pred)

print('Root mean square error :',rmse)
print('Mean Absolute error:',mae)
```

```python
def MAPE(y_true,y_pred):
    mape=np.mean(np.abs(y_true-y_pred)/y_true)
    return mape
result=MAPE(y_test,y_pred)
result###MAPE for linear Regrssion =16.22%%
```

from the above graph we can say that residuals i.e errors are randomly spread which is said to be a good fit for the linear regression model.
